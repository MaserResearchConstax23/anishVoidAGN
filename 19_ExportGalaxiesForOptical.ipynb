```python
# remove "#" at the beginning

voids_parent = read_data_db('voids_parent_colors_new_' + '04-11-23') # timestamp())
walls_parent = read_data_db('walls_parent_colors_new_' + '04-11-23') # timestamp())

voids_parent.head()
```

```python
desired_columns = [
    'object_tag',
    'Pr',
    'RA',
    'DEC',
    'PLATE',
    'FIBERID',
    'MJD'
]
```

```python
voids_parent_cols = voids_parent[desired_columns]
walls_parent_cols = walls_parent[desired_columns]

voids_parent_cols.head()
```

```python
print(str(len(voids_parent_cols)))
new_voids_filt = voids_parent_cols.dropna(axis = 'index', 
                                          how = 'any', 
                                          subset = desired_columns)
print(str(len(new_voids_filt)))

print(str(len(walls_parent_cols)))
new_walls_filt = walls_parent_cols.dropna(axis = 'index', 
                                          how = 'any', 
                                          subset = desired_columns)
print(str(len(new_walls_filt)))
```

```python
# rename old: new
new_voids_filt_named = new_voids_filt.rename(columns = {'object_tag' : 'NSAID', 'FIBERID':'FIBER'})
new_walls_filt_named = new_walls_filt.rename(columns = {'object_tag' : 'NSAID', 'FIBERID':'FIBER'})

new_voids_filt_named.head()
```
