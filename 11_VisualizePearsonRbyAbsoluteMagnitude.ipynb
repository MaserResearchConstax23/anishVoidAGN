```python
# import necessary packages

import pandas as pd # for working with tables
pd.options.mode.chained_assignment = None  # default = 'warn'

import os # for finding file directory with os.getcwd()

# for calculations
import numpy as np 
import math

import matplotlib.pyplot as plt # for plotting

# for plotting histogram with percentages
from matplotlib.ticker import PercentFormatter
import matplotlib.ticker as mtick
from matplotlib.ticker import FormatStrFormatter

from PIL import Image # for converting plots to black-and-white
from IPython.display import HTML

from ipynb.fs.full.MyFunctions_00 import *
```

```python
# read in saved Pr tables
voids_bright = read_data_db('voids_bright_' + 'new_' + '04-11-23') # timestamp())
walls_bright = read_data_db('walls_bright_' + 'new_' + '04-11-23') # timestamp())

voids_main = read_data_db('voids_main_' + 'new_' + '04-11-23') # timestamp())
walls_main = read_data_db('walls_main_' + 'new_' + '04-11-23') # timestamp())

voids_dwarf = read_data_db('voids_dwarf_' + 'new_' + '04-11-23') # timestamp())
walls_dwarf = read_data_db('walls_dwarf_' + 'new_' + '04-11-23') # timestamp())
```

```python
# Remove weird r values that occur because the object has < 3 bins for plotting

fig, ax = plt.subplots(1, 3, figsize = (18, 5), sharey = True)
plt.subplots_adjust(wspace = 0.1)

a = 0.75

num_bins = 45

# filled bars

ax[0].hist(walls_bright['Pr'],
        bins = np.linspace(-1, 1, num = num_bins, endpoint = False),
        color = "dimgray",
#         edgecolor = 'white',
#         linewidth = 1, 
        label = 'Wall galaxies',
        alpha = a,
        weights = np.ones_like(walls_bright['Pr']) / len(walls_bright['Pr'])
       ) 

ax[0].hist(voids_bright['Pr'],
        bins = np.linspace(-1, 1, num = num_bins, endpoint = False),
        color = "limegreen",
#         edgecolor = 'white',
#         linewidth = 1, 
        label = 'Void galaxies',
        alpha = a,
        weights = np.ones_like(voids_bright['Pr']) / len(voids_bright['Pr'])
       )

ax[1].hist(walls_main['Pr'],
        bins = np.linspace(-1, 1, num = num_bins, endpoint = False),
        color = "black",
#         edgecolor = 'white',
#         linewidth = 1, 
#         label = 'Wall galaxies',
        alpha = a,
        weights = np.ones_like(walls_main['Pr']) / len(walls_main['Pr'])
       ) 

ax[1].hist(voids_main['Pr'],
        bins = np.linspace(-1, 1, num = num_bins, endpoint = False),
        color = "red",
#         edgecolor = 'white',
#         linewidth = 1, 
#         label = 'Void galaxies',
        alpha = a,
        weights = np.ones_like(voids_main['Pr']) / len(voids_main['Pr'])
       )

ax[2].hist(walls_dwarf['Pr'],
        bins = np.linspace(-1, 1, num = num_bins, endpoint = False),
        color = "navy",
#         edgecolor = 'white',
#         linewidth = 1, 
        label = 'Wall galaxies',
        alpha = a,
        weights = np.ones_like(walls_dwarf['Pr']) / len(walls_dwarf['Pr'])
       ) 

ax[2].hist(voids_dwarf['Pr'],
        bins = np.linspace(-1, 1, num = num_bins, endpoint = False),
        color = "dodgerblue",
#         edgecolor = 'white',
#         linewidth = 1, 
        label = 'Void galaxies',
        alpha = a,
        weights = np.ones_like(voids_dwarf['Pr']) / len(voids_dwarf['Pr'])
       )

# ax[0].set_title('Bright sample')
# ax[1].set_title('Main sample')
# ax[2].set_title('Dwarf sample')
ax[0].text(-0.9, 0.05, 'bright voids', ha = 'left', c = 'limegreen', size = 'large')
ax[0].text(-0.9, 0.047, 'bright walls', ha = 'left', c = 'dimgray', size = 'large')

ax[1].text(-0.9, 0.05, 'main voids', ha = 'left', c = 'red', size = 'large')
ax[1].text(-0.9, 0.047, 'main walls', ha = 'left', c = 'black', size = 'large')

ax[2].text(-0.97, 0.05, 'dwarf voids', ha = 'left', c = 'dodgerblue', size = 'large')
ax[2].text(-0.97, 0.047, 'dwarf walls', ha = 'left', c = 'navy', size = 'large')

# for means and standard deviations
plot_mean_color(0, voids_bright['Pr'], walls_bright['Pr'], "limegreen", "dimgray", ax, False)
plot_mean_color(1, voids_main['Pr'], walls_main['Pr'], "red", "black", ax, False)
plot_mean_color(2, voids_dwarf['Pr'], walls_dwarf['Pr'], "dodgerblue", "navy", ax, False)

ax[0].set_ylabel('ΔN / Ntotal', size = 'x-large')
ax[1].set_xlabel('Pearson r coefficient', size = 'x-large')
# ax[1].legend(loc = 'best')

for i in range(3):
    
    ax[i].yaxis.set_major_formatter(mtick.PercentFormatter(decimals = 3))
    ax[i].set_xlim(-1.00, 1.00)
    ax[i].set_ylim(0, 0.035),
    ax[i].xaxis.set_ticks(np.arange(start = -1.0, stop = 1.1, step = 0.25))
    
    
ax[2].set_ylim(0, 0.055)
# plt.rcParams.update({'font.size': 12})

fig.savefig('Figures/iPoster-Pr-SplitByRabsmag_' + 'new_' + '.png', dpi = 300)
plt.show()
```
