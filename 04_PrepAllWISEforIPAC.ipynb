```python
# remove "#" at the beginning

new_voids_AWS = read_data('IPAC_Results/New/Voids/new_voids_AWSource')
new_walls_AWS = read_data('IPAC_Results/New/Walls/new_walls_AWSource')

new_voids_AWS.head()
```

```python
for col in new_voids_AWS.columns:
    print(col)
```

```python
ipac_new_voids = new_voids_AWS[['nsaid_01',
                             'ra',
                             'dec'
                            ]].rename(columns = {'nsaid_01' : 'nsaid'})

ipac_new_walls = new_walls_AWS[['nsaid_01',
                             'ra',
                             'dec'
                            ]].rename(columns = {'nsaid_01' : 'nsaid'})

ipac_new_voids.head()
```

```python
ipac_new_voids.loc[-1] = ['int', 'double', 'double']  # adding a row
ipac_new_voids.index = ipac_new_voids.index + 1  # shifting index
ipac_new_voids = ipac_new_voids.sort_index() 

ipac_new_voids.head(6)
```

```python
ipac_new_walls.loc[-1] = ['int', 'double', 'double']  # adding a row
ipac_new_walls.index = ipac_new_walls.index + 1  # shifting index
ipac_new_walls = ipac_new_walls.sort_index() 

ipac_new_walls.head(6)
```

```python
print(len(ipac_new_voids))
ipac_new_voids = ipac_new_voids.dropna(axis = 'index', how = 'any')
print(len(ipac_new_voids))
```

```python
print(len(ipac_new_walls))
ipac_new_walls = ipac_new_walls.dropna(axis = 'index', how = 'any')
print(len(ipac_new_walls))
```
